## Restyling a plot

```python

#Imports
import matplotlib as mpl
import matplotlib.pyplot as plt
from tessierplot import plot as ts
from tessierplot import styles

#Creating a figure + 2 horizontal subfigures
ff = plt.figure(figsize=(6.75,3.5))
gs = mpl.gridspec.GridSpec(1, 2, height_ratios=[1],width_ratios=[1, 1])

#Defining axis handles for two figures
#left plot:
ax1 = plt.subplot(gs[0, 0])
#right plot:
ax2 = plt.subplot(gs[0, 1])
gs.tight_layout(ff)

#Horizontal and vertical figure spacing
gs.update(wspace=0.50, hspace=0.5)

#Dat file to plot
file = r'./2DJVHB~9\EEJIJD~X\02VB4B~3/IvsV_J1_1V.dat'
c = ts.plotR(file)
d = ts.plotR(file)

#Plotting files with various 'styles' applied
# For available styles and their usage see the file 'styles.py' in the Tessierplot main folder
# The plot3d class takes various arguments, see the file 'plot.py' in the Tessierplot main folder
p = (c.plot3d(style=['log'], cbar_orientation='horizontal', ax_destination=ax1, value_axis=0))#, clim= [-1,50])) 
q = (d.plot3d(style=['fixlabels','mov_avg(n=2)','changeaxis(xfactor=1, xunit=K, datafactor=1e3, dataunit=k$\Omega$)','didv',
                       ], clim=[0,3],cbar_orientation='vertical', ax_destination=ax2, value_axis=1))

#Set plot axis limits
ax2.set_xlim(-5,5)
ax2.set_ylim(-5,5)
ax2.locator_params(nbins=6, axis='y')
ax2.locator_params(nbins=6, axis='x')

# Change look, size, and label of the colorbar
c.cbar.ax.set_aspect(0.15)
d.cbar.ax.set_anchor('N')
d.cbar.set_label('$\partial{I_\mathrm{S}}/\partial{VI_\mathrm{SD}}$ ($\mu$S)', y=-1.4, labelpad=-20)

# Change number of ticks on axes
d.ax.locator_params(nbins=6, axis='y')
d.ax.locator_params(nbins=6, axis='x')
c.ax.locator_params(nbins=2, axis='y')
c.ax.locator_params(nbins=2, axis='x')

# You can write on your plot
ax2.annotate( "$n=1$", xy=(0.5,0.5), horizontalalignment='right', verticalalignment='center', fontsize = 8, color='w')
ax2.annotate( "$n=2$", xy=(0.7,0.7), horizontalalignment='right', verticalalignment='center', fontsize = 8, color='w')
#close figure before saving (helps with formatting)
close()

# Saving 
ff.savefig(r'.\plots\Examplefig.png', bbox_inches = 'tight', dpi=600, transparent=True, pad_inches = 0)
ff.savefig(r'.\plots\Examplefig.pdf', bbox_inches = 'tight', dpi=600, transparent=True, pad_inches = 0)
ff.savefig(r'.\plots\Examplefig.pgf', bbox_inches = 'tight', dpi=600, transparent=True, pad_inches = 0)
#close('all')
```
